```python

import os
import torch
import numpy as np
from tqdm import tqdm
from PIL import Image
from torchvision import transforms
from torch.utils.data import DataLoader, Dataset
import matplotlib.pyplot as plt
from google.colab import drive
import zipfile

drive.mount('/content/drive')
with zipfile.ZipFile('/content/drive/MyDrive/FinalFinalFinal.zip', 'r') as zip_ref:
    zip_ref.extractall()

# Define your data directories
IMAGES_DIR_TRAIN = '/content/ADE20k_toy_dataset/images/training'
MASKS_DIR_TRAIN = '/content/ADE20k_toy_dataset/annotations/training'
IMAGES_DIR_TEST = '/content/ADE20k_toy_dataset/images/test'
MASKS_DIR_TEST = '/content/ADE20k_toy_dataset/annotations/test'
IMAGES_DIR_VALID = '/content/ADE20k_toy_dataset/images/validation'
MASKS_DIR_VALID = '/content/ADE20k_toy_dataset/annotations/validation'

# Rest of the constants
INIT_EPOCH = 0
EPOCHS = 20
BATCH_SIZE = 32
IMG_SIZE = 128
CHANNELS = 3
LEARNING_RATE = 0.0001
STEPS = 173

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('Using:', device)

def medt():
    from MedicalTransformer.lib.models.axialnet import MedT
    model = MedT(img_size = 128, imgchan = 3,num_classes=1)
    return model

class dice_loss(torch.nn.Module):
    def __init__(self):
        super(dice_loss, self).__init__()
        self.smooth = 1.

    def forward(self, logits, labels):
        logf = torch.sigmoid(logits).view(-1)
        labf = labels.view(-1)
        intersection = (logf * labf).sum()

        num = 2. * intersection + self.smooth
        den = logf.sum() + labf.sum() + self.smooth
        return 1 - (num / den)

class CustomDataset(Dataset):
    def __init__(self, images_dir, masks_dir, transform=None):
        self.images_dir = images_dir
        self.masks_dir = masks_dir
        self.transform = transform
        self.image_filenames = sorted(os.listdir(images_dir))
        self.mask_filenames = sorted(os.listdir(masks_dir))

    def __len__(self):
        return len(self.image_filenames)

    def __getitem__(self, idx):
        image_filename = self.image_filenames[idx]
        mask_filename = self.mask_filenames[idx]

        image_path = os.path.join(self.images_dir, image_filename)
        mask_path = os.path.join(self.masks_dir, mask_filename)

        image = Image.open(image_path).convert('RGB')
        mask = Image.open(mask_path).convert('L')

        if self.transform:
            image = self.transform(image)
            mask = self.transform(mask)

        return image, mask

def visualize_dataset(loader, num_images=5):
    fig, axes = plt.subplots(2, num_images, figsize=(15, 5))
    for i, (image, mask) in enumerate(loader):
        if i >= num_images:
            break

        axes[0, i].imshow(image[0].permute(1, 2, 0))  # Display the first channel of the image (assuming it's RGB)
        axes[0, i].set_title('Image')

        axes[1, i].imshow(mask[0][0], cmap='gray')  # Display the mask
        axes[1, i].set_title('Mask')

        axes[0, i].axis('off')
        axes[1, i].axis('off')

    plt.tight_layout()
    plt.show()

def get_data_generators():
    train_transforms = transforms.Compose([
        transforms.Resize((IMG_SIZE, IMG_SIZE)),
        transforms.ToTensor()
    ])

    val_transforms = transforms.Compose([
        transforms.Resize((IMG_SIZE, IMG_SIZE)),
        transforms.ToTensor(),
    ])

    train_dataset = CustomDataset(images_dir=IMAGES_DIR_TRAIN, masks_dir=MASKS_DIR_TRAIN, transform=train_transforms)
    val_dataset = CustomDataset(images_dir=IMAGES_DIR_VALID, masks_dir=MASKS_DIR_VALID, transform=val_transforms)
    test_dataset = CustomDataset(images_dir=IMAGES_DIR_TEST, masks_dir=MASKS_DIR_TEST, transform=val_transforms)

    train_loader = DataLoader(dataset=train_dataset, batch_size=BATCH_SIZE, shuffle=False)
    val_loader = DataLoader(dataset=val_dataset, batch_size=BATCH_SIZE, shuffle=False)
    test_loader = DataLoader(dataset=test_dataset, batch_size=BATCH_SIZE, shuffle=False)

    return train_loader, val_loader,test_loader

train_loader,val_loader,test_loader  = get_data_generators()
visualize_dataset(train_loader)

import numpy as np
import torch
from tqdm import tqdm
import pandas as pd

import numpy as np
import torch
from tqdm import tqdm
import pandas as pd

def train(model, criterion, opt, scheduler):
    global INIT_EPOCH, EPOCHS, BATCH_SIZE, device, SAVE_PATH

    model.train()

    train_accuracy_values = []
    train_precision_values = []
    train_iou_values = []
    train_loss_values = []  # Initialize list for training loss

    val_accuracy_values = []
    val_precision_values = []
    val_iou_values = []
    val_loss_values = []  # Initialize list for validation loss

    for epoch in range(INIT_EPOCH, EPOCHS):
        train_loader, val_loader, test_loader = get_data_generators()
        total_accuracy_train = []
        total_precision_train = []
        total_iou_train = []

        total_accuracy_val = []
        total_precision_val = []
        total_iou_val = []

        total_loss_train = 0.0
        total_loss_val = 0.0

        # Training phase
        model.train()
        for batch in tqdm(train_loader):
            images, masks = batch
            images = images.to(device)
            masks = masks.to(device)

            opt.zero_grad(set_to_none=True)

            outputs = model(images)
            loss = criterion(outputs, masks)

            loss.backward()
            opt.step()
            scheduler.step()

            # Calculate metrics
            with torch.no_grad():
                preds = torch.sigmoid(outputs) > 0.5
                true_positives = (preds & masks.byte()).sum()
                predicted_positives = preds.sum()
                actual_positives = masks.sum()
                precision = true_positives / (predicted_positives + 1e-8)
                iou = true_positives / (actual_positives + predicted_positives - true_positives + 1e-8)

                total_accuracy_train.append(1 - loss.item())
                total_precision_train.append(precision.item())
                total_iou_train.append(iou.item())

                # Accumulate the training loss
                total_loss_train += loss.item()

        # Validation phase
        model.eval()
        with torch.no_grad():
            for batch in tqdm(val_loader):
                images, masks = batch
                images = images.to(device)
                masks = masks.to(device)

                outputs = model(images)

                # Calculate metrics
                preds = torch.sigmoid(outputs) > 0.5
                true_positives = (preds & masks.byte()).sum()
                predicted_positives = preds.sum()
                actual_positives = masks.sum()
                loss = criterion(outputs, masks)
                precision = true_positives / (predicted_positives + 1e-8)
                iou = true_positives / (actual_positives + predicted_positives - true_positives + 1e-8)

                total_accuracy_val.append(1 - loss.item())  # Adjust this if you have a separate validation loss
                total_precision_val.append(precision.item())
                total_iou_val.append(iou.item())

                # Accumulate the validation loss
                total_loss_val += loss.item()

        # Calculate the average training and validation loss
        avg_train_loss = total_loss_train / len(train_loader)
        avg_val_loss = total_loss_val / len(val_loader)

        train_loss_values.append(avg_train_loss)  # Append the training loss to the list
        val_loss_values.append(avg_val_loss)  # Append the validation loss to the list

        train_accuracy_values.append(np.mean(total_accuracy_train))
        train_precision_values.append(np.mean(total_precision_train))
        train_iou_values.append(np.mean(total_iou_train))

        val_accuracy_values.append(np.mean(total_accuracy_val))
        val_precision_values.append(np.mean(total_precision_val))
        val_iou_values.append(np.mean(total_iou_val))

        print('Epoch: {}/{} - Training: accuracy: {:.4f} - precision: {:.4f} - IoU: {:.4f} - Loss: {:.4f} | Validation: accuracy: {:.4f} - precision: {:.4f} - IoU: {:.4f} - Loss: {:.4f}'.format(
            epoch + 1, EPOCHS,
            np.mean(total_accuracy_train), np.mean(total_precision_train), np.mean(total_iou_train), avg_train_loss,
            np.mean(total_accuracy_val), np.mean(total_precision_val), np.mean(total_iou_val), avg_val_loss
        ))

        if (epoch + 1) % 5 == 0:
            torch.save({'epoch': epoch + 1,
                        'model_state_dict': model.state_dict(),
                        'optimizer_state_dict': opt.state_dict(),
                        'loss': avg_val_loss,  # Save validation loss as checkpoint
                        }, SAVE_PATH)
            print('Checkpoint saved.')

    # Save metric values as CSV file
    metrics_data = {
        'Train_Accuracy': train_accuracy_values,
        'Train_Precision': train_precision_values,
        'Train_IoU': train_iou_values,
        'Train_Loss': train_loss_values,  # Add training loss to metrics
        'Validation_Accuracy': val_accuracy_values,
        'Validation_Precision': val_precision_values,
        'Validation_IoU': val_iou_values,
        'Validation_Loss': val_loss_values  # Add validation loss to metrics
    }

    metrics_df = pd.DataFrame(metrics_data)
    metrics_df.to_csv('metrics.csv', index=False)
    return model


# Set your desired model name
name = 'trans'
print("I am here")
model = medt().to(device)
SAVE_PATH = './model_checkpoint_transUnet_latest.pt'

# Initialize optimizer and scheduler
criterion = dice_loss()
opt = torch.optim.AdamW(model.parameters(), lr=LEARNING_RATE)
STEPS = len(train_loader)
scheduler = torch.optim.lr_scheduler.OneCycleLR(opt,
                                                max_lr=LEARNING_RATE * 10,
                                                steps_per_epoch=STEPS,
                                                pct_start=0.15,
                                                epochs=EPOCHS
                                                )

# Call the train function to start training
history = train(model, criterion, opt, scheduler)
```

```python
import torch
import torchvision.transforms.functional as TF
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score

def evaluate_and_visualize(model, dataloader, num_images=20):
    model.eval()

    total_iou = 0.0
    total_dice = 0.0
    total_samples = 0
    all_true_labels = []
    all_pred_labels = []

    with torch.no_grad():
        for batch_idx, (images, masks) in enumerate(tqdm(dataloader)):
            images = images.to(device)
            masks = masks.to(device)

            outputs = model(images)
            preds = torch.sigmoid(outputs) > 0.5

            intersection = (preds & masks.byte()).sum().item()
            union = (preds | masks.byte()).sum().item()

            iou = intersection / (union + 1e-8)
            dice = 2 * intersection / (preds.sum().item() + masks.sum().item() + 1e-8)

            total_iou += iou * images.size(0)  # Accumulate for the whole batch
            total_dice += dice * images.size(0)  # Accumulate for the whole batch
            total_samples += images.size(0)

            # Flatten masks and predictions for classification metrics
            true_labels = masks.view(-1).cpu().numpy()
            pred_labels = preds.view(-1).cpu().numpy()

            all_true_labels.extend(true_labels)
            all_pred_labels.extend(pred_labels)

            if batch_idx == 0:
                visualization_images = images[:num_images].cpu()
                visualization_masks = masks[:num_images].cpu()
                visualization_preds = preds[:num_images].cpu()

    avg_iou = total_iou / total_samples
    avg_dice = total_dice / total_samples

    f1 = f1_score(all_true_labels, all_pred_labels)
    precision = precision_score(all_true_labels, all_pred_labels)
    recall = recall_score(all_true_labels, all_pred_labels)
    accuracy = accuracy_score(all_true_labels, all_pred_labels)

    print('IoU on test data: {:.4f}'.format(avg_iou))
    print('Dice Coefficient on test data: {:.4f}'.format(avg_dice))
    print('F1 Score on test data: {:.4f}'.format(f1))
    print('Precision on test data: {:.4f}'.format(precision))
    print('Recall on test data: {:.4f}'.format(recall))
    print('Accuracy on test data: {:.4f}'.format(accuracy))

    # Visualize the selected images, masks, and predictions
    for idx in range(num_images):
        plt.figure(figsize=(10, 6))

        plt.subplot(1, 3, 1)
        plt.imshow(TF.to_pil_image(visualization_images[idx].squeeze()))
        plt.title('Image')

        plt.subplot(1, 3, 2)
        plt.imshow(visualization_masks[idx].squeeze(), cmap='gray')
        plt.title('Mask')

        plt.subplot(1, 3, 3)
        plt.imshow(visualization_preds[idx].squeeze(), cmap='gray')
        plt.title('Prediction')

        plt.tight_layout()
        plt.show()

# Call the evaluate_and_visualize function after training is complete
train_loader, val_loader, test_loader = get_data_generators()
evaluate_and_visualize(model, test_loader)
```

```python

import os
import torch
import numpy as np
from tqdm import tqdm
from PIL import Image
from torchvision import transforms
from torch.utils.data import DataLoader, Dataset
from self_attention_cv.transunet import TransUnet
import matplotlib.pyplot as plt
from google.colab import drive
import zipfile

drive.mount('/content/drive')
with zipfile.ZipFile('/content/drive/MyDrive/FinalFinalFinal.zip', 'r') as zip_ref:
    zip_ref.extractall()

# Define your data directories
IMAGES_DIR_TRAIN = '/content/ADE20k_toy_dataset/images/training'
MASKS_DIR_TRAIN = '/content/ADE20k_toy_dataset/annotations/training'
IMAGES_DIR_TEST = '/content/ADE20k_toy_dataset/images/test'
MASKS_DIR_TEST = '/content/ADE20k_toy_dataset/annotations/test'
IMAGES_DIR_VALID = '/content/ADE20k_toy_dataset/images/validation'
MASKS_DIR_VALID = '/content/ADE20k_toy_dataset/annotations/validation'

# Rest of the constants
INIT_EPOCH = 0
EPOCHS = 60
BATCH_SIZE = 64
IMG_SIZE = 128
CHANNELS = 3
LEARNING_RATE = 1e-4
training_image_len = len([f for f in os.listdir('/content/ADE20k_toy_dataset/images/training')])
remaining_data = training_image_len % BATCH_SIZE  # Calculate the remaining data that won't fit in full batches
STEPS = (training_image_len // BATCH_SIZE) + (1 if remaining_data > 0 else 0)
print(training_image_len,'total_trainign_image')
print(STEPS,'-----------------------STEPS--------------')
print(remaining_data,'remaining_data')

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('Using:', device)

def trans_unet(image_size, channels=3, blocks=8, embeddings=512, classes=1):
    from self_attention_cv.transunet import TransUnet
    return TransUnet(in_channels=channels,
                     img_dim=image_size,
                     classes=classes)

class dice_loss(torch.nn.Module):
    def __init__(self):
        super(dice_loss, self).__init__()
        self.smooth = 1.

    def forward(self, logits, labels):
        logf = torch.sigmoid(logits).view(-1)
        labf = labels.view(-1)
        intersection = (logf * labf).sum()

        num = 2. * intersection + self.smooth
        den = logf.sum() + labf.sum() + self.smooth
        return 1 - (num / den)

class CustomDataset(Dataset):
    def __init__(self, images_dir, masks_dir, transform=None):
        self.images_dir = images_dir
        self.masks_dir = masks_dir
        self.transform = transform
        self.image_filenames = sorted(os.listdir(images_dir))
        self.mask_filenames = sorted(os.listdir(masks_dir))

    def __len__(self):
        return len(self.image_filenames)

    def __getitem__(self, idx):
        image_filename = self.image_filenames[idx]
        mask_filename = self.mask_filenames[idx]

        image_path = os.path.join(self.images_dir, image_filename)
        mask_path = os.path.join(self.masks_dir, mask_filename)

        image = Image.open(image_path).convert('RGB')
        mask = Image.open(mask_path).convert('L')

        if self.transform:
            image = self.transform(image)
            mask = self.transform(mask)

        return image, mask

def visualize_dataset(loader, num_images=5):
    fig, axes = plt.subplots(2, num_images, figsize=(15, 5))
    for i, (image, mask) in enumerate(loader):
        if i >= num_images:
            break

        axes[0, i].imshow(image[0].permute(1, 2, 0))  # Display the first channel of the image (assuming it's RGB)
        axes[0, i].set_title('Image')

        axes[1, i].imshow(mask[0][0], cmap='gray')  # Display the mask
        axes[1, i].set_title('Mask')

        axes[0, i].axis('off')
        axes[1, i].axis('off')

    plt.tight_layout()
    plt.show()

def get_data_generators():
    train_transforms = transforms.Compose([
        transforms.Resize((IMG_SIZE, IMG_SIZE)),
        transforms.ToTensor()
    ])

    val_transforms = transforms.Compose([
        transforms.Resize((IMG_SIZE, IMG_SIZE)),
        transforms.ToTensor(),
    ])

    train_dataset = CustomDataset(images_dir=IMAGES_DIR_TRAIN, masks_dir=MASKS_DIR_TRAIN, transform=train_transforms)
    val_dataset = CustomDataset(images_dir=IMAGES_DIR_VALID, masks_dir=MASKS_DIR_VALID, transform=val_transforms)
    test_dataset = CustomDataset(images_dir=IMAGES_DIR_TEST, masks_dir=MASKS_DIR_TEST, transform=val_transforms)

    train_loader = DataLoader(dataset=train_dataset, batch_size=BATCH_SIZE, shuffle=False)
    val_loader = DataLoader(dataset=val_dataset, batch_size=BATCH_SIZE, shuffle=False)
    test_loader = DataLoader(dataset=test_dataset, batch_size=BATCH_SIZE, shuffle=False)

    return train_loader, val_loader,test_loader

train_loader,val_loader,test_loader  = get_data_generators()
visualize_dataset(train_loader)

import numpy as np
import torch
from tqdm import tqdm
import pandas as pd

import numpy as np
import torch
from tqdm import tqdm
import pandas as pd

def train(model, criterion, opt, scheduler):
    global INIT_EPOCH, EPOCHS, BATCH_SIZE, device, SAVE_PATH

    model.train()

    train_accuracy_values = []
    train_precision_values = []
    train_iou_values = []
    train_loss_values = []  # Initialize list for training loss

    val_accuracy_values = []
    val_precision_values = []
    val_iou_values = []
    val_loss_values = []  # Initialize list for validation loss

    for epoch in range(INIT_EPOCH, EPOCHS):
        train_loader, val_loader, test_loader = get_data_generators()

        total_accuracy_train = []
        total_precision_train = []
        total_iou_train = []

        total_accuracy_val = []
        total_precision_val = []
        total_iou_val = []

        total_loss_train = 0.0
        total_loss_val = 0.0

        # Training phase
        model.train()
        for batch in tqdm(train_loader):
            images, masks = batch
            images = images.to(device)
            masks = masks.to(device)

            opt.zero_grad(set_to_none=True)

            outputs = model(images)
            loss = criterion(outputs, masks)

            loss.backward()
            opt.step()
            scheduler.step()

            # Calculate metrics
            with torch.no_grad():
                preds = torch.sigmoid(outputs) > 0.5
                true_positives = (preds & masks.byte()).sum()
                predicted_positives = preds.sum()
                actual_positives = masks.sum()
                precision = true_positives / (predicted_positives + 1e-8)
                iou = true_positives / (actual_positives + predicted_positives - true_positives + 1e-8)

                total_accuracy_train.append(1 - loss.item())
                total_precision_train.append(precision.item())
                total_iou_train.append(iou.item())

                # Accumulate the training loss
                total_loss_train += loss.item()

        # Validation phase
        model.eval()
        with torch.no_grad():
            for batch in tqdm(val_loader):
                images, masks = batch
                images = images.to(device)
                masks = masks.to(device)

                outputs = model(images)

                # Calculate metrics
                preds = torch.sigmoid(outputs) > 0.5
                true_positives = (preds & masks.byte()).sum()
                predicted_positives = preds.sum()
                actual_positives = masks.sum()
                loss = criterion(outputs, masks)
                precision = true_positives / (predicted_positives + 1e-8)
                iou = true_positives / (actual_positives + predicted_positives - true_positives + 1e-8)

                total_accuracy_val.append(1 - loss.item())  # Adjust this if you have a separate validation loss
                total_precision_val.append(precision.item())
                total_iou_val.append(iou.item())

                # Accumulate the validation loss
                total_loss_val += loss.item()

        # Calculate the average training and validation loss
        avg_train_loss = total_loss_train / len(train_loader)
        avg_val_loss = total_loss_val / len(val_loader)

        train_loss_values.append(avg_train_loss)  # Append the training loss to the list
        val_loss_values.append(avg_val_loss)  # Append the validation loss to the list

        train_accuracy_values.append(np.mean(total_accuracy_train))
        train_precision_values.append(np.mean(total_precision_train))
        train_iou_values.append(np.mean(total_iou_train))

        val_accuracy_values.append(np.mean(total_accuracy_val))
        val_precision_values.append(np.mean(total_precision_val))
        val_iou_values.append(np.mean(total_iou_val))

        print('Epoch: {}/{} - Training: accuracy: {:.4f} - precision: {:.4f} - IoU: {:.4f} - Loss: {:.4f} | Validation: accuracy: {:.4f} - precision: {:.4f} - IoU: {:.4f} - Loss: {:.4f}'.format(
            epoch + 1, EPOCHS,
            np.mean(total_accuracy_train), np.mean(total_precision_train), np.mean(total_iou_train), avg_train_loss,
            np.mean(total_accuracy_val), np.mean(total_precision_val), np.mean(total_iou_val), avg_val_loss
        ))

        if (epoch + 1) % 5 == 0:
            torch.save({'epoch': epoch + 1,
                        'model_state_dict': model.state_dict(),
                        'optimizer_state_dict': opt.state_dict(),
                        'loss': avg_val_loss,  # Save validation loss as checkpoint
                        }, SAVE_PATH)
            print('Checkpoint saved.')

    # Save metric values as CSV file
    metrics_data = {
        'Train_Accuracy': train_accuracy_values,
        'Train_Precision': train_precision_values,
        'Train_IoU': train_iou_values,
        'Train_Loss': train_loss_values,  # Add training loss to metrics
        'Validation_Accuracy': val_accuracy_values,
        'Validation_Precision': val_precision_values,
        'Validation_IoU': val_iou_values,
        'Validation_Loss': val_loss_values  # Add validation loss to metrics
    }

    metrics_df = pd.DataFrame(metrics_data)
    metrics_df.to_csv('metrics.csv', index=False)
    return model


# Set your desired model name
name = 'trans'
print("I am here")
model = trans_unet(IMG_SIZE).to(device)
SAVE_PATH = './model_checkpoint_transUnet_latest.pt'

# Initialize optimizer and scheduler
criterion = dice_loss()
opt = torch.optim.AdamW(model.parameters(), lr=LEARNING_RATE)
scheduler = torch.optim.lr_scheduler.OneCycleLR(opt,
                                                max_lr=1e-4,
                                                steps_per_epoch=STEPS,
                                                pct_start=0.15,
                                                epochs=EPOCHS
                                                )

# Call the train function to start training
history = train(model, criterion, opt, scheduler)
```

```python
import torch
import torchvision.transforms.functional as TF
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score

def evaluate_and_visualize(model, dataloader, num_images=20):
    model.eval()

    total_iou = 0.0
    total_dice = 0.0
    total_samples = 0
    all_true_labels = []
    all_pred_labels = []

    with torch.no_grad():
        for batch_idx, (images, masks) in enumerate(tqdm(dataloader)):
            images = images.to(device)
            masks = masks.to(device)

            outputs = model(images)
            preds = torch.sigmoid(outputs) > 0.5

            intersection = (preds & masks.byte()).sum().item()
            union = (preds | masks.byte()).sum().item()

            iou = intersection / (union + 1e-8)
            dice = 2 * intersection / (preds.sum().item() + masks.sum().item() + 1e-8)

            total_iou += iou * images.size(0)  # Accumulate for the whole batch
            total_dice += dice * images.size(0)  # Accumulate for the whole batch
            total_samples += images.size(0)

            # Flatten masks and predictions for classification metrics
            true_labels = masks.view(-1).cpu().numpy()
            pred_labels = preds.view(-1).cpu().numpy()

            all_true_labels.extend(true_labels)
            all_pred_labels.extend(pred_labels)

            if batch_idx == 0:
                visualization_images = images[:num_images].cpu()
                visualization_masks = masks[:num_images].cpu()
                visualization_preds = preds[:num_images].cpu()

    avg_iou = total_iou / total_samples
    avg_dice = total_dice / total_samples

    f1 = f1_score(all_true_labels, all_pred_labels)
    precision = precision_score(all_true_labels, all_pred_labels)
    recall = recall_score(all_true_labels, all_pred_labels)
    accuracy = accuracy_score(all_true_labels, all_pred_labels)

    print('IoU on test data: {:.4f}'.format(avg_iou))
    print('Dice Coefficient on test data: {:.4f}'.format(avg_dice))
    print('F1 Score on test data: {:.4f}'.format(f1))
    print('Precision on test data: {:.4f}'.format(precision))
    print('Recall on test data: {:.4f}'.format(recall))
    print('Accuracy on test data: {:.4f}'.format(accuracy))

    # Visualize the selected images, masks, and predictions
    for idx in range(num_images):
        plt.figure(figsize=(10, 6))

        plt.subplot(1, 3, 1)
        plt.imshow(TF.to_pil_image(visualization_images[idx].squeeze()))
        plt.title('Image')

        plt.subplot(1, 3, 2)
        plt.imshow(visualization_masks[idx].squeeze(), cmap='gray')
        plt.title('Mask')

        plt.subplot(1, 3, 3)
        plt.imshow(visualization_preds[idx].squeeze(), cmap='gray')
        plt.title('Prediction')

        plt.tight_layout()
        plt.show()

# Call the evaluate_and_visualize function after training is complete
train_loader, val_loader, test_loader = get_data_generators()
evaluate_and_visualize(model, test_loader)
```
